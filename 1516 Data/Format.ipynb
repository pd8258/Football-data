```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
out1 = pd.read_csv('FantraxPlayers_Outfield.csv')
out2 = pd.read_csv('FantraxPlayers_Outfield2.csv')
gk = pd.read_csv('FantraxPlayers_Goalkeepers.csv')
```

```python
out1 = pd.read_csv('FantraxPlayers_Outfield (1).csv')
out2 = pd.read_csv('FantraxPlayers_Outfield2 (1).csv')
gk = pd.read_csv('FantraxPlayers_Goalkeepers (1).csv')
```

```python
out_1 = out1.drop(out1.columns[4:7], axis=1)
del out_1['Rk']
out_2 = out2.drop(out2.columns[4:7], axis=1)
del out_2['Rk']
gks = gk.drop(gk.columns[4:7], axis=1)
del gks['Rk']
```

```python
out = pd.merge(out_1,out_2)
full = pd.concat([out,gks],ignore_index=True)
```

```python
players = full['Player']
names = []
teams = []
for player in players:
      if player[-3:] == 'OUR':
        names.append(player[:-7])
        teams.append(player[-4:])
      else:
        names.append(player[:-6])
        teams.append(player[-3:])
full['Name'] = names
full['Team'] = teams
    
indexed = full.set_index(['Team','Name'])
```

```python
indexed.to_csv('Week1.csv')
```
